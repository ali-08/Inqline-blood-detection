```python
df.columns
```

```python
length = df['X'].count()
length
```

```python
rate = 0.0
for i in range(0,length-1):
    rate = rate + (df.loc[i+1,'X'] / df.loc[i, 'X'])
rate = rate / (df.loc[:,"X"].count()-1)
               
rate
```

```python
for i in range(1,length):
    df.loc[i, "Y"] = df.loc[i-1, "X"] * rate *(df.loc[i,"Days_per_Month"] / df.loc[i-1,"Days_per_Month"])
    
```

```python
df[["X", "Y"]]
```

```python
df.dropna(inplace=True)
```

### MAE is not accurate because data values are too big  

```python

from sklearn.metrics import mean_squared_error

lin_mse = mean_squared_error(df[["X"]], df[["Y"]])
lin_rmse = np.sqrt(lin_mse)
print(lin_rmse)
```

### USING MAPE

```python
mape = np.mean(np.abs((df["X"] - df["Y"]) / df["X"])) * 100
```
